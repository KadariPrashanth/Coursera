```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
import seaborn as sns
%matplotlib notebook
plt.style.use('seaborn-colorblind')
```

```python
df1 = pd.read_csv('MinTemp.csv',index_col ='YEAR')
df1=df1.rename(columns={'ANNUAL':'AvgMin'})
df1.head()
```

```python
df2 = pd.read_csv('datafile2.csv',index_col ='YEAR')#,usecols = ['YEAR','OCT-DEC']
df2=df2.rename(columns={'ANNUAL':'AvgMax'})
df2.head()
```

```python
plt.figure()
ax3 = plt.subplot()

ax3.plot(df2['AvgMax'],'y', alpha =0.75,  label = 'Avg IND Max Temperature in Degrees Celsius')
ax3.plot(df1['AvgMin'],'r', alpha =0.75,  label = 'Avg IND Min Temperature in Degrees Celsius')

ax3.set_xlabel('Year')
ax3.set_ylabel('Temperature in Degrees Celsius')
plt.title('Yearly Average Max vs Min Temperatures in India from 1901-2011')
plt.tight_layout()
ax3.legend(['AvgMax','AvgMin'], loc=1, frameon=True, prop={'size': 5})
```

```python
plt.figure()
ax2 = plt.subplot()

ax2.plot(df2['JAN-FEB'], 'g', label = 'Avg IND Max Temp of JAN-FEB in Degrees Celsius')
ax2.plot(df2['MAR-MAY'], 'k', label = 'Avg IND Max Temp of MAR-MAY in Degrees Celsius')
ax2.plot(df2['JUN-SEP'], 'r', label = 'Avg IND Max Temp of JUN-SEP in Degrees Celsius')
ax2.plot(df2['OCT-DEC'], 'y', label = 'Avg IND Max Temp of OCT-DEC in Degrees Celsius')


ax2.set_xlabel('Year')
ax2.set_ylabel('Temperature in Degrees Celsius')
plt.title('Seasonal Average Max Temperatures in India from 1901 - 2011')
plt.tight_layout()
ax2.legend(['JAN-FEB','MAR-MAY','JUN-SEP','OCT-DEC'], loc=1, frameon=True, prop={'size': 5})
```

```python
plt.figure()
ax = plt.subplot()

ax.plot(df1['JAN-FEB'], 'g', label = 'Avg IND Min Temp of JAN-FEB in Degrees Celsius')
ax.plot(df1['MAR-MAY'], 'k', label = 'Avg IND Min Temp of MAR-MAY in Degrees Celsius')
ax.plot(df1['JUN-SEP'], 'r', label = 'Avg IND Min Temp of JUN-SEP in Degrees Celsius')
ax.plot(df1['OCT-DEC'], 'y', label = 'Avg IND Min Temp of OCT-DEC in Degrees Celsius')

ax.set_xlabel('Year')
ax.set_ylabel('Temperature in Degrees Celsius')
plt.title('Seasonal Average Min Temperatures in India from 1901 - 2011')
plt.tight_layout()
ax.legend(['JAN-FEB','MAR-MAY','JUN-SEP','OCT-DEC'], loc=1, frameon=True, prop={'size': 5})
```

```python
df3=df1
df3=df3.drop(['JAN-FEB', 'MAR-MAY', 'JUN-SEP','OCT-DEC'], axis=1)
df3

df4=df2
df4=df4.drop(['JAN-FEB', 'MAR-MAY', 'JUN-SEP','OCT-DEC'], axis=1)
df4

df5 = df3.join(df4)
df5.head()
```

```python
#from pandas.plotting import scatter_matrix
pd.tools.plotting.scatter_matrix(df5, figsize=(10,10), alpha=0.5)
```

```python
cor = sns.pairplot(df5, diag_kind='kde', size=4);
plt.subplots_adjust(top=0.9)
cor.fig.suptitle('Correlationship : India Min VS Max Temp from 1901-2011')
```
